```python
import pandas as pd
bd = pd.read_csv("./trip.csv",error_bad_lines=False)
bd.head()
```

1. What is the average trip duration for a borrowed bicycle?

```python
bd['tripduration'].mean()
```

2. What’s the most common age of a bicycle-sharer?

```python
bd['age'] = 2021 - bd['birthyear']
bd['age'].value_counts().head()
```

3. Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
wd = pd.read_csv('./weather.csv')
wd.head()
```

```python
import numpy as np
wd['month'] = wd['Date'].astype('datetime64[ns]')
wd['month'] = pd.DatetimeIndex(wd['month']).month
wd.groupby("month")["Precipitation_In"].agg([np.mean,np.median])
```

4. What’s the average number of bikes at a given bike station?

```python
bd['day'] = bd['starttime'].astype('datetime64[ns]')
bd['day'] = pd.DatetimeIndex(bd['day']).day
bd.groupby('from_station_id')['day'].mean()
```
